```python
import pandas as pd
import os
import re
```

##### Leer el archivo UNA_SOLITARIA_VOZ_HUMANA.txt con with open
##### Convertir a minúsculas
#####  Conservar sólo palabras
##### Tokenizar

```python
with open("VOZ.txt", encoding = 'utF-8') as archivo1: 
    UNAVOZ = archivo1.read()
```

```python

UNAVOZ = UNAVOZ.lower() # Minúscula
UNAVOZ = re.sub(r"[\W\d_]+", " ", UNAVOZ) # Dejar sólo letras
UNAVOZ = UNAVOZ.strip() # Quitar espacios del principio
UNAVOZ = UNAVOZ.split() # Tokenización
UNAVOZ
```

##### Quitar palabras vacias
##### ¿Cuáles son las 10 palabras no vacias más usadas?

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')
```

```python

UNAVOZ_ = [palabra for palabra in UNAVOZ if palabra not in stopwords_sp]
UNAVOZ_
```

```python
#frecuenciaPalab = []
#for w in UNAVOZ_:
#    frecuenciaPalab.append(UNAVOZ_.count(w))
from collections import Counter
cnt = Counter()
for word in UNAVOZ_:
    cnt[word] += 1
cnt
df = pd.DataFrame([[key, cnt[key]] for key in cnt.keys()], columns=['Palabra', 'Repeticion'])
DF=df.sort_values('Repeticion', ascending= False)
DF.head(10)
```

#### Stemming del documento
#### ¿Cuáles son las 10 raíces más usadas?

```python
from nltk.stem.snowball import SnowballStemmer
englishStemmer=SnowballStemmer("english")
spanishStemmer=SnowballStemmer("spanish")
```

```python
texto_stem=[]

for palabra in UNAVOZ_:
    raiz = spanishStemmer.stem(palabra)
    texto_stem.append(raiz)
    
texto_stem
```

```python
CNT_STEM = Counter()
for word in texto_stem:
    CNT_STEM[word] += 1

df_stem = pd.DataFrame([[key, CNT_STEM[key]] for key in CNT_STEM.keys()], columns=['Palabra', 'Repeticion'])
DF=df_stem.sort_values('Repeticion', ascending= False)
DF.head(10)
```

#### Contar cuántas ocurrencias hay por cada parte de la oración

```python
import es_core_news_sm

nlp = es_core_news_sm.load()

b=""
for palabra in UNAVOZ_:
   b = b + str(palabra)+ " "
doc = nlp(b)

palab=[(w.text, w.pos_) for w in doc]

palab2=pd.DataFrame(palab)
palab2.columns = ['Palabra','Elemento']
palab2.groupby(['Elemento']).size().reset_index(name='cuenta').sort_values('cuenta', ascending=False)
```
